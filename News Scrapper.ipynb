```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time

# Function to set up the driver
def setup_driver():
    options = Options()
    options.headless = True  # Running in headless mode
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    return driver
```

## Yahoo Finance
//a[@class="tw-w-full titles-link noUnderline svelte-wdkn18"] - Main Head

//a[@class="subtle-link fin-size-small titles noUnderline svelte-wdkn18"] - Sub Head

```python
# Function to scrape Yahoo Finance news
def scrape_yahoo_finance_news():
    driver = setup_driver()
    data = []  # List to store headlines
    try:
        # URL of the Yahoo Finance news section
        url = 'https://finance.yahoo.com/'
        driver.get(url)
        time.sleep(5)  # Wait for the page to load

        # Finding all the news headlines based on given XPaths
        headlines = driver.find_elements(By.XPATH, '//a[@class="tw-w-full titles-link noUnderline svelte-wdkn18"]') + \
                    driver.find_elements(By.XPATH, '//a[@class="subtle-link fin-size-small titles noUnderline svelte-wdkn18"]')
        for headline in headlines:
            data.append(headline.text)  # Adding headline text to the list

    finally:
        driver.quit()

    # Creating a DataFrame
    yahoo_finance_df = pd.DataFrame(data, columns=['Headline'])
    return yahoo_finance_df

# Run the scraper function and print the DataFrame
yahoo_finance_df = scrape_yahoo_finance_news()

yahoo_finance_df
```

```python
def scrape_bloomberg_news():
    driver = setup_driver()
    data = []  # List to store headlines

    try:
        url = 'https://www.bloomberg.com/markets'
        driver.get(url)
        articles = driver.find_elements(By.XPATH, '//div[@data-type="article"]')
        for article in articles:
            try:
                headline_element = article.find_element(By.XPATH, './/div[@data-component="headline"]')
                data.append(headline_element.text)
            except:
                continue  # Skip if headline element is not found

    finally:
        driver.quit()

    bloomberg_df = pd.DataFrame(data, columns=['Headline'])
    return bloomberg_df

# Run the scraper function and print the DataFrame
bloomberg_df = scrape_bloomberg_news()

bloomberg_df
```
