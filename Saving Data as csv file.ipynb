```python
import pandas as pd
from pymongo import MongoClient

# Setup MongoDB connection
client = MongoClient("mongodb://localhost:27017/")
db = client["news_articles_db"]

# Get all collection names in the database
collection_names = db.list_collection_names()

# Iterate over each collection and export data to CSV
for collection_name in collection_names:
    # Load collection data into a DataFrame
    collection = db[collection_name]
    data = list(collection.find())
    
    # Convert to DataFrame, dropping the MongoDB ID
    df = pd.DataFrame(data).drop('_id', axis=1)
    
    # Save to CSV with the collection name
    csv_filename = f"{collection_name}_news_data.csv"
    df.to_csv(csv_filename, index=False)
    print(f"Data from {collection_name} saved to {csv_filename}")
```

```python
import pandas as pd
import glob

# Define the path where your CSV files are stored
# For example, if your files are in a folder named 'news_articles' on your desktop
path = '/Users/priyakundu/Documents/NYU Projects/Research - Carlos/News and Prices Correlation/News Data'  # Replace with the path to your CSV files
all_files = glob.glob(path + "/*.csv")

# Create an empty DataFrame to store all the data
all_articles = pd.DataFrame()

# Loop through all the files and read each one into a DataFrame
for filename in all_files:
    df = pd.read_csv(filename)
    # Assuming 'DateTime' is the column name containing the datetime information
    # Convert 'DateTime' to datetime type if not already
    df['DateTime'] = pd.to_datetime(df['DateTime'])
    # Extract date and time into separate columns
    df['Date'] = df['DateTime'].dt.date
    df['Time'] = df['DateTime'].dt.time
    # Append to the main DataFrame
    all_articles = pd.concat([all_articles, df], ignore_index=True)

# Drop the original 'DateTime' column if no longer needed
all_articles.drop('DateTime', axis=1, inplace=True)

# Sort the DataFrame by the 'Date' and 'Time' columns
all_articles.sort_values(by=['Date', 'Time'], inplace=True)

# Save the combined data to a new CSV file
all_articles.to_csv("news_data.csv", index=False)
```
